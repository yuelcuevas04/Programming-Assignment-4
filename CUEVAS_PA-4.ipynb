```python
import pandas as pd
```

```python
eceboards = pd.read_excel('board2.xlsx')
eceboards
```

```python
Instru = eceboards.loc[(eceboards['Track']=='Instrumentation') & (eceboards['Hometown']=='Luzon') & (eceboards['Electronics'] >70),['Name','GEAS','Electronics']]
print(Instru)
```

```python
eceboards['Average'] = eceboards[['Math','Electronics','GEAS','Communication']].mean(axis=1)
```

```python
Mindy = eceboards.loc[(eceboards['Hometown']=='Mindanao') & (eceboards['Gender']=='Female') & (eceboards['Average']>=55),['Name','Track','Electronics','Average']]
print(Mindy)
```

```python
import matplotlib.pyplot as plt
avg = eceboards.groupby('Track')['Average'].mean()
plt.bar (avg.index, avg.values)
plt.ylabel('Mean') 
plt.xlabel('Tracks') 
plt.show()
```

```python
avg = eceboards.groupby('Gender')['Average'].mean()
plt.bar (avg.index, avg.values)
plt.ylabel('Mean') 
plt.xlabel('Gender') 
plt.show()
```

```python
avg = eceboards.groupby('Hometown')['Average'].mean()
plt.bar (avg.index, avg.values)
plt.ylabel('Mean') 
plt.xlabel('Hometown') 
plt.show()
```
